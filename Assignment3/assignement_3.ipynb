# Different Visualisation on iris dataset
 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("iris.csv")
df
```

```python
print("First 5 rows of the iris dataset:")
print(df.head())
```

```python
print("Any null/missing values? ")
print(df.isnull().sum())
```

```python
print("Dataset Info:")
print(df.info())
```

#  Line Graph (Sepal Length over Samples)



The line graph shows how Sepal Length varies across all 150 iris samples. The pattern appears smooth overall, with noticeable fluctuations between samples of different species.


```python
plt.figure(figsize=(10, 5))
plt.plot(df.index, df['SepalLengthCm'], color='blue', marker='o')
plt.title('Sepal Length Across Samples')
plt.xlabel('Sample Index')
plt.ylabel('Sepal Length (cm)')
plt.grid(True)
plt.show()
```

# Histogram (Petal Width Distribution)

The histogram reveals that most flowers have a Petal Width between 0.1 cm and 0.5 cm. The data is not uniformly distributed, indicating some species have much thicker petals than others.

```python
plt.figure(figsize=(8, 5))
plt.hist(df['PetalWidthCm'], bins=15, color='green', edgecolor='black')
plt.title('Distribution of Petal Width')
plt.xlabel('Petal Width (cm)')
plt.ylabel('Frequency')
plt.show()
```

# Pie Chart (Species Count)

All three speciesâ€”Setosa, Versicolor, and Virginicaâ€”are equally represented in the dataset, each comprising 33.3% of the total samples (50 each out of 150).

```python
species_counts = df['Species'].value_counts()
plt.figure(figsize=(9, 9))
plt.pie(species_counts, labels=species_counts.index, autopct='%1.1f%%')
plt.title('Distribution of Iris Species')
plt.show()
```

# Box & Whisker Plot (Sepal Width by Species)

Petal Length and Petal Width show a strong positive correlation (~0.96), meaning as one increases, so does the other. Sepal Length also shows moderate correlation with petal features, while Sepal Width has a weaker relationship.

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='Species', y='SepalWidthCm', data=df, palette='pastel')
plt.title('Sepal Width Distribution by Species')
plt.xlabel('Species')
plt.ylabel('Sepal Width (cm)')
plt.show()
```

# Heatmap (Correlation Matrix)

```python
plt.figure(figsize=(8, 6))
corr_matrix = df.drop('Species', axis=1).corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Feature Correlation Heatmap')
plt.show()
```
